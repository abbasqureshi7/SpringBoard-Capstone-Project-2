```python
import re
import string
import nltk
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Function to remove twitter handle
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
        
    return input_txt
```

```python
#Remove twitter handles
tweet_df['tidy_tweet'] = np.vectorize(remove_pattern)(tweet_df['Tweet'], "@[\w]*")
```

```python
tweet_df.head()
```

```python
#Removing short words from Tweets
tweet_df['tidy_tweet'] = tweet_df['tidy_tweet'].apply(lambda x: ' '.join ([w for w in x.split() if len(w)>3]))
```

```python
tweet_df.head()
```

```python
#Tokenizing Tweets
tokenized_tweet = tweet_df['tidy_tweet'].apply(lambda x: x.split())
tokenized_tweet.head()
```

```python
#Stemming the words, removing the suffixes from words.
from nltk.stem.porter import *
stemmer = PorterStemmer()

tokenized_tweet = tokenized_tweet.apply(lambda x: [stemmer.stem(i) for i in x]) #this will do the stemming
tokenized_tweet.head()
```

```python
#Remove http URLs
tweet_df['tidy_tweet'] = re.sub(r"http\S+", "", str)
```
